```python
import pandas as pd
import numpy as np
from tqdm import tqdm

tqdm.pandas()
```

```python
df_olympics = pd.read_csv('data/olympic_history/athlete_events.csv')
df_noc = pd.read_csv('data/olympic_history/noc_regions.csv')
```

```python
df_olympics.head()
```

```python
df_olympics['Medal'] = df_olympics['Medal'].fillna("")
df_olympics.loc[df_olympics['Medal'] == 'Gold', 'mcount'] = 3
df_olympics.loc[df_olympics['Medal'] == 'Silver', 'mcount'] = 2
df_olympics.loc[df_olympics['Medal'] == 'Bronze', 'mcount'] = 1
df_olympics.loc[df_olympics['Medal'] == '', 'mcount'] = 0

df_olympics['mcount'] = df_olympics['mcount'].astype(int)

df_olympics.head()
```

```python
df_olympics = df_olympics.groupby(['NOC', 'Year']).agg({'mcount': 'sum'}).reset_index()
```

```python
df_olympics = df_olympics.merge(df_noc, on='NOC')

df_olympics.head()
```

```python
df_olympics = df_olympics.drop(['notes'], axis = 1)
```

```python
df_olympics.columns = ['NOC', 'Year', 'mcount', 'Country']
df_olympics = df_olympics.dropna()
df_olympics.head()
```

```python
df_population = pd.read_csv('data/population_over_years.csv')
df_population['NOC'] = df_population['Country Code']
```

```python
display(df_population.head())
display(df_olympics.head())
```

```python
df_over_gdp = pd.read_csv('data/gdp_over_years.csv')
df_over_gdp['NOC'] = df_over_gdp['Country Code']
```

```python
def retrieve_gdp_for_year(row):
    try:
        year = str(row['Year'])
        if year in df_over_gdp.columns and year in df_population.columns:
            row['gdp'] = df_over_gdp.loc[df_over_gdp['NOC'] == row['NOC'], year].sum()
            row['population'] = df_population.loc[df_population['NOC'] == row['NOC'], year].sum()
        return row
    except:
        print(row)
        return row

final_dataset = df_olympics.copy()
final_dataset = final_dataset.progress_apply(lambda row: retrieve_gdp_for_year(row), axis=1)
print(final_dataset.shape)
final_dataset = final_dataset.dropna()
print(final_dataset.shape)

print(min(final_dataset['Year']))
print(max(final_dataset['Year']))
```

```python
df_regions = pd.read_csv('data/countries of the world.csv')
```

```python
def retrieve_region(row):
    try:
        country = row['Country']
        match = df_regions.loc[df_regions['Country'].str.match(country), 'Region'].unique()
        if len(match) > 0:
            row['Region'] = match[0]
        return row
    except:
        raise
    
final_dataset = final_dataset.progress_apply(lambda row: retrieve_region(row), axis=1)
print(final_dataset.shape)
final_dataset = final_dataset.dropna()
print(final_dataset.shape)
```

```python
from geopy.geocoders import Nominatim
latlng_country = {}

countries = final_dataset['Country'].copy().unique()

geolocator = Nominatim(user_agent="data-processing")
for country in countries:
    try:
        loc = geolocator.geocode(country)
        latlng_country[country] = (loc.latitude, loc.longitude)
    except:
        print("unable to geocode " + country)
```

```python
def geolocate(row):
    country = row['Country']
    try:
        loc = latlng_country[country]
        row['lat'] = loc[0]
        row['long'] = loc[1]
        return row
    except:
        return np.nan

final_dataset = final_dataset.progress_apply(lambda row: geolocate(row), axis=1)
print(final_dataset.shape)
final_dataset = final_dataset.dropna()
print(final_dataset.shape)
```

```python
display(final_dataset)
```

```python
final_dataset.columns = ['Country', 'NOC', 'Region', 'Year', 'Per Capita GDP', 'Total Medals', 'Population', 'lat', 'lon']
```

```python
display(final_dataset)
```

```python
final_dataset = (final_dataset[np.logical_and(final_dataset['Per Capita GDP'] != 0, final_dataset['Population'] != 0)])
```

```python
final_dataset.to_csv('data/final_data.csv')
```
